# Conclusiones, recomendaciones y acciones de negocio

### 🔹 Tipo de contrato
**Conclusión:**  
Los clientes con contratos mensuales tienen mayor probabilidad de abandonar. En cambio, los contratos a dos años están asociados a una menor tasa de abandono.

**Recomendación:**  
Fomentar la contratación de planes a largo plazo.

**Acción de negocio:**  
Diseñar campañas promocionales que ofrezcan beneficios exclusivos (descuentos, servicios gratuitos, puntos de fidelidad) para quienes contraten planes anuales o bianuales.

---

### 🔹 Número de llamadas al servicio al cliente  
**Conclusión:**  
A mayor número de llamadas al soporte, mayor es la probabilidad de churn. Especialmente a partir de la tercera llamada.

**Recomendación:**  
Mejorar la eficiencia y resolución en las primeras interacciones con el cliente.

**Acción de negocio:**  
Implementar un sistema de seguimiento para identificar clientes que han llamado más de dos veces y asignarles soporte especializado para evitar la pérdida.

---

### 🔹 Plan familiar  
**Conclusión:**  
Los clientes con planes familiares tienen menor tasa de abandono. Compartir el servicio parece generar mayor compromiso.

**Recomendación:**  
Incentivar la contratación de planes familiares.

**Acción de negocio:**  
Lanzar campañas dirigidas con descuentos grupales, posibilidad de añadir miembros fácilmente o beneficios compartidos (como almacenamiento extra o atención prioritaria).

---

### 🔹 Aplicación de descuentos  
**Conclusión:**  
Los descuentos aplicados no parecen reducir significativamente el abandono. Se podría estar perdiendo dinero en descuentos que no influyen en la fidelización del cliente.

**Recomendación:**  
Evaluar su efectividad o redirigir estas estrategias hacia otras iniciativas que generen mayor compromiso con el servicio

**Acción de negocio:**  
Redirigir los recursos de descuentos hacia acciones más eficaces, como recompensas por lealtad o mejoras en el servicio.

---

### 🔹 Servicios adicionales (protección y respaldo)  
**Conclusión:**  
Los clientes que contratan servicios adicionales como protección de dispositivos o backup en la nube tienden a abandonar menos.

**Recomendación:**  
Potenciar los servicios complementarios como estrategia de retención.

**Acción de negocio:**  
Ofrecer paquetes combinados con descuentos si se contratan servicios adicionales, destacando su valor añadido.

---

### 🔹 Duración de la cuenta  
**Conclusión:**  
Los clientes nuevos (con menor tiempo en la plataforma) abandonan con más facilidad.

**Recomendación:**  
Fortalecer la etapa de onboarding para nuevos clientes.

**Acción de negocio:**  
Diseñar una experiencia de bienvenida con tutoriales, beneficios iniciales y seguimiento personalizado durante los primeros meses.

---

### 🔹 Clientes menores de 30 años  
**Conclusión:**  
Presentan una mayor fidelidad al servicio.

**Recomendación:**  
Aprovechar esta afinidad generacional para consolidar la relación con la marca.

**Acción de negocio:**  
Desarrollar campañas enfocadas a jóvenes con ventajas tecnológicas, precios especiales para estudiantes o contenido personalizado.

---

### 🔹 Clientes mayores de 65 años  
**Conclusión:**  
Tienen una tasa de abandono significativamente más alta.

**Recomendación:**  
Detectar posibles barreras en usabilidad y experiencia de usuario.

**Acción de negocio:**  
Crear una interfaz simplificada, soporte asistido o formación personalizada para mayores, y adaptar el contenido a sus preferencias.

---

### 🔹 Método de pago  
**Conclusión:**  
Los pagos manuales (como con tarjeta de débito) están asociados a una mayor tasa de abandono.

**Recomendación:**  
Facilitar y promover los métodos de pago automáticos.

**Acción de negocio:**  
Ofrecer incentivos por activar pagos automáticos, como un mes gratis, prioridad en soporte o menor cargo por servicio.

---

### 🔹 Cargo mensual  
**Conclusión:**  
A mayor cargo mensual, el abandono tiende a aumentar, aunque este patrón se revierte a partir de los 60 dólares.

**Recomendación:**  
Equilibrar la percepción de valor en los planes con cargos altos.

**Acción de negocio:**  
Aumentar el valor percibido en los planes premium mediante beneficios exclusivos y campañas que expliquen mejor sus ventajas.
